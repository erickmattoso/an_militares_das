# scrapping pep pessoas politicamente expostas
Esse notebook tem por objetivo fazer uma coleta de dados do [PORTAL DA TRANSPARÊNCIA](http://www.portaltransparencia.gov.br/).<br>
Os dados aqui buscados são informações em formato `.CSV` das __pessoas politicamente expostas.__

```python
from multiprocessing.pool import ThreadPool
from urllib.request import urlretrieve
import requests
import urllib.request
```

```python
def download_url(url):
    file_name = url.split('/')[-1]    
    r = requests.get(url, stream=True)
    if r.status_code == requests.codes.ok:
        print('downloading: ',file_name,'\n')
        urllib.request.urlretrieve(url, 'pep/' + file_name +'.zip')
    else:
        print('Sorry, url not found.')
```

```python
urls = []
for ano in range(12,20):
    for mes in ["%.2d" % mes for mes in range(1,13)]:
        url = 'http://www.portaltransparencia.gov.br/download-de-dados/pep/20' + str(ano) + str(mes)
        urls.append(url)
urls[:2]
```

```python
results = ThreadPool(5).imap_unordered(download_url, urls)
for r in results:
    print('results = ', r)
```
