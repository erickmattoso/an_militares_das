# scrapping servidores militares
Esse notebook tem por objetivo fazer uma coleta de dados do [PORTAL DA TRANSPARÊNCIA](http://www.portaltransparencia.gov.br/).<br>
Os dados aqui buscados são informações em formato `.CSV` dos __servidores públicos militares do Brasil.__

```python
%run '../libs.py'
```

```python
path = 'servidores_militares'
os.mkdir(path)
```

```python
def download_url(url):
    file_name = url.split('/')[-1]
    r = requests.get(url, stream=True)
    if r.status_code == requests.codes.ok:
        print('downloading: ', file_name, '\n')
        urllib.request.urlretrieve(url, path + '/' + file_name + '.zip')
    else:
        print('Sorry, url not found.')
```

```python
urls = []
for ano in range(13, 20):
    for mes in ["%.2d" % mes for mes in range(1, 13)]:
        url = 'http://www.portaltransparencia.gov.br/download-de-dados/servidores/20' + \
            str(ano) + str(mes) + '_Militares'
        urls.append(url)
```

```python
results = ThreadPool(5).imap_unordered(download_url, urls)
for r in results:
    print('results = ', r)
```
